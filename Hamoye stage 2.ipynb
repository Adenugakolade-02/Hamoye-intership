```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import os
os.chdir(r'C:\Users\user\Downloads\UDEMY\datasets\hamoye 2')
train=pd.read_csv('energydata_complete___appliances_energy_prediction_data_set_.csv')
```

```python
train.head()
```

```python
train = train.drop(['date','lights'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
train[train.dtypes.index.to_list()] = MinMaxScaler().fit_transform(train[train.dtypes.index.to_list()])
```

```python
train.head()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
prediction=train['Appliances']
del(train['Appliances'])
```

```python
#question1

x=np.array(train['T2']).reshape(-1,1)
y=np.array(train['T6']).reshape(-1,1)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
hmodel = LinearRegression()
hmodel.fit(X_train,y_train)
pred = hmodel.predict(X_test)
score = r2_score(pred,y_test)
print(score)
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,prediction,test_size=0.3,random_state=42)
```

```python
model_lin=LinearRegression()
model_lin.fit(X_train,y_train)
predicted=model_lin.predict(X_test)
```

```python
#question 13
print('MAE_score',mean_absolute_error(predicted,y_test)) 
```

```python
#question 14
print('RSS',sum((predicted-y_test)**2))
```

```python
#question 15
print('RMSE_score',np.sqrt(mean_squared_error(predicted,y_test)))
```

```python
#question 16
print('r2_score',r2_score(predicted,y_test))
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
#question 17
linear_model_weights = get_weights_df(model_lin,X_train,'Linear')
linear_model_weights
```

```python
#question 18
model_rid=Ridge(alpha=0.4)
model_rid.fit(X_train,y_train)
predicted=model_rid.predict(X_test)
print('RMSE_score',np.sqrt(mean_squared_error(predicted,y_test)))
```

```python
#question 19
model_las=Lasso(alpha=0.001)
model_las.fit(X_train,y_train)
predicted=model_las.predict(X_test)
lasso_weights_df = get_weights_df(model_las,X_train,'Lasso')
lasso_weights_df
```

```python
#question 20
print('RMSE_score',np.sqrt(mean_squared_error(predicted,y_test)))
```
